```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error

from linear_regression import LinearRegression
```

```python
df1 = pd.read_csv('../data1.csv')
df2 = pd.read_csv('../data2.csv')
df3 = pd.read_csv('../data3.csv')
```

```python
# data1
x_data1 = df1['x1'].values.reshape(-1, 1)
y_data1 = df1['x2'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data1, y_data1, test_size=0.3, random_state=0)
```

```python
plt.scatter(X_train, y_train, label='obserbed')
```

```python
plt.scatter(X_test, y_test, label='obserbed')
```

```python
digrees = [1, 3, 5, 10, 15]
```

```python
for d in digrees:
    model = LinearRegression(digree=d)
    model.fit(X_train, y_train)
    
    pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, pred)
    print('digree = {}: {}'.format(d, mse))
```

```python
# data1
model_1 = LinearRegression(digree=3)
X = df1['x1'].values.reshape(-1, 1)
y = df1['x2'].values
model_1.fit(X, y)
reg_eq_1 = model_1.get_equation()

w = model_1.coef
linex = np.linspace(-5, 5)
y_ = model_1.predict(linex.reshape(-1, 1))

plt.figure(figsize=(5, 5))
plt.scatter(X, y, label='obserbed')
plt.plot(linex, y_, c='indianred', label=reg_eq_1)
plt.xlim(-6, 6)
plt.ylim(-12, 14)
plt.legend()
plt.savefig('data1.png')
```

```python
# data2
x_data2 = df2['x1'].values.reshape(-1, 1)
y_data2 = df2['x2'].values
X_train, X_test, y_train, y_test = train_test_split(x_data2, y_data2, test_size=0.3, random_state=0)
```

```python
digrees = [1, 3, 5, 10, 15]
```
